# MH 时序关系

```python
import numpy as np
from tqdm import tqdm
import os
import numpy as np
import matplotlib.pyplot as plt
folder_path = './cpp_data_conv_times/'  # 替换为你的文件夹路径
# 使用 listdir 获取文件夹中所有文件的文件名
file_names = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

num_mh = 2000
str_mh = "MH"
times = []
bate0s = []
bate1s = []
sigmas = []

with open(folder_path+"/"+str_mh+"_beta0_"+str(num_mh)+".txt", 'r') as file:
    for line in file:
        bate0s.append(float(line.strip()))
    bate0s = np.array(bate0s)
with open(folder_path+"/"+str_mh+"_beta1_"+str(num_mh)+".txt", 'r') as file:
    for line in file:
        bate1s.append(float(line.strip()))
    bate1s = np.array(bate1s)
with open(folder_path+"/"+str_mh+"_sigma_"+str(num_mh)+".txt", 'r') as file:
    for line in file:
        sigmas.append(float(line.strip()))
    sigmas = np.array(sigmas)
with open(folder_path+"/"+str_mh+"_time"+str(num_mh)+".txt", 'r') as file:
    for line in file:
        times.append(float(line.strip()))
    times = np.array(times)
    times = times *1000

pars = np.vstack((bate0s,bate1s,sigmas))
np.save(folder_path+"/"+str_mh+'_pars.npy',pars) 
np.save(folder_path+"/"+str_mh+'_times.npy',times)


```

```python
np.vstack((bate0s,bate1s,sigmas)).shape
```

```python
plt.plot(times,bate0s)
plt.plot(times,bate1s)
plt.plot(times,sigmas)
plt.xlabel("Time: ms")

plt.legend()
```

```python
iteration = range(bate0s.shape[0])
plt.plot(iteration,bate0s)
plt.plot(iteration,bate1s)
plt.plot(iteration,sigmas)
plt.xlabel("Iteration")
```

# 综合对比

```python
class data:
    def __init__(self,pars_path, times_path):
        self.pars = np.load(pars_path) 
        self.times = np.load(times_path) 

MH_pars_path = "./cpp_data_conv_times/MH_pars.npy"
MH_times_path = "./cpp_data_conv_times/MH_times.npy"
MH_data = data(MH_pars_path,MH_times_path)

MP_pars_path = "./cpp_data_conv_times/MP_pars.npy"
MP_times_path = "./cpp_data_conv_times/MP_times.npy"
MP_data = data(MP_pars_path,MP_times_path)

PMP_pars_path = "./cpp_data_conv_times/PMP_pars.npy"
PMP_times_path = "./cpp_data_conv_times/PMP_times.npy"
PMP_data = data(PMP_pars_path,PMP_times_path)
```

```python
colors = 'bgr'
pars= ["$\\beta_0$", "$\\beta_1$", "$\\epsilon$"]

sample_num = 1500
line = np.linspace(0, sample_num, sample_num)
for i, j, par in zip(range(MH_data.pars.shape[0]), colors, pars):
    plt.plot(line, MH_data.pars[i, :sample_num], label="SP "+ par, color=j, linestyle="-")


line = np.linspace(0, sample_num, sample_num)
for i, j, par in zip(range(MH_data.pars.shape[0]), colors, pars):
    plt.plot(line, MP_data.pars[i, :sample_num], label="MP "+ par, color=j, linestyle=":")


line = np.linspace(0, sample_num, sample_num)
for i, j, par in zip(range(MH_data.pars.shape[0]), colors, pars):
    plt.plot(line, PMP_data.pars[i, :sample_num], label="PMP "+ par, color=j, linestyle="-.")

plt.legend()
plt.xlabel("Number of iterations")
plt.ylabel("values of parameters")
```

```python
colors = 'bgr'
pars= ["$\\beta_0$", "$\\beta_1$", "$\\epsilon$"]
sample_num = 1500
for i, j, par in zip(range(MH_data.pars.shape[0]), colors, pars):
    plt.plot(MH_data.times[:sample_num], MH_data.pars[i, :sample_num], label="SP "+ par, color=j, linestyle="-")

for i, j, par in zip(range(MH_data.pars.shape[0]), colors, pars):
    plt.plot(MP_data.times[:sample_num], MP_data.pars[i, :sample_num], label="MP "+ par, color=j, linestyle=":")

for i, j, par in zip(range(MH_data.pars.shape[0]), colors, pars):
    plt.plot(PMP_data.times[:sample_num], PMP_data.pars[i, :sample_num], label="PMP "+ par, color=j, linestyle="-.")

plt.legend()

plt.xlabel("time: ms")
plt.ylabel("values of parameters")
```

```python
fig, ax = plt.subplots(2,1,figsize=(10, 12))

sample_num = 1900
line = np.linspace(0, sample_num, sample_num)
for i, j, par in zip(range(MH_data.pars.shape[0]), colors, pars):
    ax[0].plot(line, MH_data.pars[i, :sample_num], label="SP "+ par, color=j, linestyle="-")


line = np.linspace(0, sample_num, sample_num)
for i, j, par in zip(range(MH_data.pars.shape[0]), colors, pars):
    ax[0].plot(line, MP_data.pars[i, :sample_num], label="MP "+ par, color=j, linestyle=":")


line = np.linspace(0, sample_num, sample_num)
for i, j, par in zip(range(MH_data.pars.shape[0]), colors, pars):
    ax[0].plot(line, PMP_data.pars[i, :sample_num], label="PMP "+ par, color=j, linestyle="-.")

ax[0].legend()
ax[0].set_xlabel("Number of iterations")
ax[0].set_ylabel("values of parameters")
ax[0].set_xlim(-0,1000)

for i, j, par in zip(range(MH_data.pars.shape[0]), colors, pars):
    ax[1].plot(MH_data.times[:], MH_data.pars[i, :], label="SP "+ par, color=j, linestyle="-")

for i, j, par in zip(range(MH_data.pars.shape[0]), colors, pars):
    ax[1].plot(MP_data.times[:], MP_data.pars[i, :], label="MP "+ par, color=j, linestyle=":")

for i, j, par in zip(range(MH_data.pars.shape[0]), colors, pars):
    ax[1].plot(PMP_data.times[:], PMP_data.pars[i, :], label="PMP "+ par, color=j, linestyle="-.")

ax[1].legend()
ax[1].set_xlim(-0,90000)
ax[1].set_xlabel("time: ms")
ax[1].set_ylabel("values of parameters")
plt.savefig("./conv_iter_and_time.pdf")
```
